```python
import numpy as np

def manual_dilation(image, SE):
    
    rows, cols = image.shape
    se_rows, se_cols = SE.shape
    pad_h, pad_w = se_rows // 2, se_cols // 2 
    
    output = np.zeros_like(image)

    padded_img = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='constant')

    for i in range(rows):
        for j in range(cols):
            neighborhood = padded_img[i:i+se_rows, j:j+se_cols]
            
        
            output[i, j] = np.max(neighborhood[SE == 1])
            
    return output 
```
